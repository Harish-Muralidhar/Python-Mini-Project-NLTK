```python
# Importing required libraries
import nltk
from nltk.tokenize import sent_tokenize
import math
```

```python
# Reading the file
f = open("Article.txt")
article=f.read()
```

```python
# Removing certain characters to clean the paragraph
preprocessing=["Â\xa0","(",")","\"","â€”","/","-","[","]","|",":"]
for i in preprocessing:
    article=article.replace(i," ")
article=article.replace("W.F.","W F ")
```

```python
# list of sentences
sentences=sent_tokenize(article)
```

```python
# Computing lenght of list which is the number of sentences in the paragraph
len(sentences)
```

```python
# Computing dictionary of dictionaries
wordfreq = {}
dicts={}
for sentence in sentences:
    sentence=sentence.lower()
    words=sentence.split()
    d = {x:words.count(x) for x in words}
    wordfreq={sentence:d}
    dicts.update(wordfreq)
```

```python
# displaying the dictionary
dicts
```

```python
# computing term frequency
tf=[]
for key,value in dicts.items():
    print("\n")
    print(key)
    total=sum(value.values())
    print("\nThe term frequencies of the above sentence\n")
    for k,v in value.items():
        print(k,"-",(v/total))
        tf.append(v/total)
```

```python
# displaying term frequency
tf
```

```python
# Generating term frequency matrix such that value of each term is its term frequency
dicts_termfreq=dicts
i=0
for key,value in dicts_termfreq.items():
    for k,v in value.items():
        dicts_termfreq[key][k]=tf[i]
        i=i+1
```

```python
# displaying the term frequency matrix as a dictionary
dicts_termfreq
```

```python
# Computing set of words from the given article
word_list=[]
for sentence in sentences:
    sentence=sentence.lower()
    words=sentence.split()
    for i in words:
        word_list.append(i)
words_set=list(set(word_list))
dicts_count={}
for word in words_set:
    count=0
    for sentence in sentences:
        sentence=sentence.lower()
        split=sentence.split()
        if word in split:
            count=count+1
    temp={word:count}
    dicts_count.update(temp)
```

```python
# displaying the dictionary with the word as key and the number of sentences it is in as values
dicts_count
```

```python
# computing idf for each word
words_set
total_number_of_sentences=len(sentences)
dicts_idf={}
for word in words_set:
    idf=math.log(total_number_of_sentences/dicts_count[word])
    temp={word:idf}
    dicts_idf.update(temp)
```

```python
# displaying idf of each word
dicts_idf
```

```python
# computing tf-idf
dict_tf_idf={}
tf_idf=[]
for sentence in sentences:
    i=0
    sentence=sentence.lower()
    split=sentence.split()
    for w in split:
        tf_idf.append(dicts_termfreq[sentence][w]*dicts_idf[w])
```

```python
# creating a dictionary to represent tf-idf of each word in each sentence
dicts_tf_idf=dicts
i=0
for key,value in dicts_tf_idf.items():
    for k,v in value.items():
        dicts_tf_idf[key][k]=tf_idf[i]
        i=i+1
```

```python
# displaying tf-idf of each word in each sentence
dicts_tf_idf
```

```python
# finding weights
weights=[]
dicts_weights={}
sentence_length=[]
for sentence in sentences:
    sentence_length.append(len(sentence.split()))
i=0
for key,value in dicts_tf_idf.items():
    weights.append(sum(value.values())/sentence_length[i])
```

```python
# creating dictionary of weights for each sentence
j=0
for sentence in sentences:
    temp={sentence:weights[j]}
    j=j+1
    dicts_weights.update(temp)
```

```python
# displaying dictionary of weights
dicts_weights
```

```python
# Computing threshold
threshold=sum(dicts_weights.values())/len(dicts_weights.values())
```

```python
# displaying threshold
threshold
```

```python
# computing summary from weights and count of number of sentences in summary
summary=" "
summary_count=0
for key,value in dicts_weights.items():
    if value>threshold:
        summary=summary+key
        summary_count=summary_count+1
```
